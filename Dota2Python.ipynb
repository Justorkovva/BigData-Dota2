```python
# Zadaniem jest sprawdzić czy wynik jest funkcją pozostałych zmiennych, w związku z czym nie będzie występował w tej macierzy
# Tryb gry i region mógłby być brany pod uwagę, ale w tym projekcie chciałabym się skupić tylko na kwestii wyboru bohaterów
# Dane opisujące tryb gry i mapę pozwoliły za to sprawdzić poprawność rekordów i ich przydatność do projektu

# z poniższych instrukcji na razie rezygnuję, ze względu na połączenie plików csv
#y = dota2train.iloc[:, 4:]
#y.head()
#X = dota2train.iloc[:, 4:]
#X.head()
#y_test = dota2test.iloc[:, 4:]
#y_test.head()
#X_test = dota2test.iloc[:, 4:]
#X_test.head()

#Przygotowanie danych
X = dota2.iloc[:, 4:]
X.head()
```

```python
#zadanie polega na określeniu jaki będzie wynik
y = dota2["Wynik"]
y[0:10]
```

```python
#Podział zbioru na próbę uczącą i testową 
#Zamiast pierwotnego podziału 90%-10%, mogę teraz zadać inny np. 80%-20% 
import sklearn.model_selection
np.arange(4)
np.arange(X.shape[0])

idx_ucz, idx_test = sklearn.model_selection.train_test_split(np.arange(X.shape[0]), test_size=0.2, random_state=12345)
X_ucz, X_test = X.iloc[idx_ucz, :], X.iloc[idx_test, :]
y_ucz, y_test = y[idx_ucz], y[idx_test]
X_ucz.shape, X_test.shape, y_ucz.shape, y_test.shape
```

```python
#Funkcja z internetu, żeby usłyszeć kiedy dany blok się skończy wykonywać
from IPython.display import Audio, display

def allDone():
  display(Audio(url='https://sound.peal.io/ps/audios/000/000/537/original/woo_vu_luvub_dub_dub.wav', autoplay=True))
```

```python
#metoda k-najbliższych sąsiadów
import sklearn.neighbors
knn = sklearn.neighbors.KNeighborsClassifier(n_neighbors=1000, p=1) 
knn.fit(X_ucz, y_ucz)
#parametry klasyfikatora
allDone()
```

```python
#predykcja która drużyna wygra grę
yk_pred = knn.predict(X_test)
allDone()
```

```python
#ocena jakości klasyfikatora
#np. procent poprawnie zaklasyfikowanych obserwacji
import sklearn.metrics
sklearn.metrics.accuracy_score(y_test, yk_pred)
```

```python
#macierz pomyłek [[true negative, false positive], [false negative, true positive]]
sklearn.metrics.confusion_matrix(y_test, yk_pred)
```
